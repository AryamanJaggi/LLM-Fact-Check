```python
#device = torch.device("cuda") # whisperx does not work unless you use a gpu
device = "cuda"
batch_size = 16 # standard. Was not working until I swithced bathc size to 4 for some reason but now working with 16. I think the problem was I was trying to downgrade torch when I shouldnt have
compute_type = "float16"
#device, compute_type=compute_type
model = whisperx.load_model("large-v3", device, compute_type=compute_type, language='en')
#model = whisperx.load_model("large-v2",device, compute_type=compute_type, language='en') # maybe use v3 instead

#INFO:pytorch_lightning.utilities.migration.utils:Lightning automatically upgraded your loaded checkpoint from v1.5.4 to v2.2.0. To apply the upgrade to your files permanently, run `python -m pytorch_lightning.utilities.upgrade_checkpoint ../root/.cache/torch/whisperx-vad-segmentation.bin`
#Model was trained with pyannote.audio 0.0.1, yours is 3.1.1. Bad things might happen unless you revert pyannote.audio to 0.x.
#Model was trained with torch 1.10.0+cu102, yours is 2.0.0+rocm5. Bad things might happen unless you revert torch to 1.x.
```

```python
HuggingFaceToken = ''
#Temporary for singular court tv video
file_path_audio = '/content/drive/My Drive/YoutubeTranscriptions_Truth/final_video_3.mp4'
file_path_transcription = '/content/drive/My Drive/YoutubeTranscriptions_Truth/text1.txt'

audio = whisperx.load_audio(file_path_audio)

#transcribe audio
result = model.transcribe(audio, batch_size=batch_size)

# 2. Align whisper output
model_a, metadata = whisperx.load_align_model(language_code="en", device=device)
result = whisperx.align(result["segments"], model_a, metadata, audio, device, return_char_alignments=False)


# 3. Diarize audio
diarize_model = whisperx.DiarizationPipeline(use_auth_token=HuggingFaceToken, device=device)
diarize_segments = diarize_model(audio)


# 4. Merge segments
result = whisperx.assign_word_speakers(diarize_segments, result)

with open(file_path_transcription,"w", encoding="utf-8") as f:
     for item in result["segments"]:
        # Safely access dictionary values with .get() to avoid KeyError
        start = str(item.get("start", "Unknown start time"))
        end = str(item.get("end", "Unknown end time"))
        speaker = item.get("speaker", "Unknown speaker")
        text = item.get("text", "No text available")

        # Write formatted string to file
        f.write(f"time: {start} - {end}\n{speaker}: {text}\n")
```
